1] What is a Decision Tree, and how does it work?

A Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks. It models decisions and their possible consequences as a tree-like structure.

Working Mechanism:

The root node starts with the entire dataset.

At each internal node, the algorithm finds the best feature and threshold to split the dataset based on a splitting criterion (like Gini or Entropy).

This process continues recursively, forming branches and leaf nodes.

Each leaf node represents a final decision or prediction:

For classification: a class label.

For regression: a numerical value (mean of samples).

The core idea is to partition the data space in a way that increases class purity at every split.

2] What are impurity measures in Decision Trees?

Impurity measures are metrics used to determine how mixed the classes are at a node. A node is pure if all data points belong to one class.

Most common impurity measures:

Gini Impurity (used in CART)

Entropy (used in ID3/C4.5)

Variance (used in regression trees)

These measures help evaluate how "good" a feature split is ‚Äî i.e., which split results in the most homogeneous child nodes.

6] What is the difference between Gini Impurity and Entropy?

Gini Impurity and Entropy are both measures of node impurity in decision trees, used to evaluate the quality of a split. Gini Impurity is a simpler, faster computation that focuses on the probability of misclassification. Entropy, while slightly more complex to calculate, tends to create more balanced trees.
Key Differences:
Computation:
Gini Impurity is faster to compute as it involves simpler calculations than Entropy, which uses logarithms.
Tree Balance:
Entropy tends to create more balanced trees, while Gini Impurity might favor isolating the most frequent class in a separate branch.
Range:
Gini Impurity ranges from 0 to 0.5, while Entropy ranges from 0 to 1.
Performance:
While both perform similarly in practice, some studies suggest that Entropy might yield slightly better results, especially in smaller datasets where time complexity is less of a concern.
In Summary:
Gini Impurity is computationally cheaper and faster to calculate.
Entropy tends to create more balanced trees.
The choice between Gini Impurity and Entropy depends on factors like dataset size, time constraints, and desired tree balance. 7] What is the mathematical explanation behind Decision Trees?

Decision Trees use mathematical metrics like Gini Impurity or Information Gain to determine the best way to split data into subsets. These metrics quantify the "purity" or homogeneity of the data within a node, guiding the algorithm to find splits that maximize the separation of different classes.
Here's a more detailed breakdown:
1. Gini Impurity:

Definition:

Gini Impurity measures the impurity of a node, with a value of 0 indicating a perfectly homogeneous node (all samples belong to the same class) and 1 indicating maximal impurity (equal distribution of classes).

Calculation:

It's calculated as 1 - (probability of class 1)^2 - (probability of class 2)^2 - ... for all classes. For example, if a node has 2 samples of class A and 3 samples of class B, the Gini Impurity would be 1 - (2/5)^2 - (3/5)^2 = 0.48.

Split Evaluation:

When evaluating a split, the Gini Impurity of the resulting child nodes is calculated and weighted by the number of samples in each child. The split that leads to the lowest overall weighted Gini Impurity is chosen as the best split.

2. Information Gain:

Definition:

Information Gain measures how much the entropy (randomness) of the dataset decreases after a split.

Calculation:

It's calculated as the entropy of the parent node minus the weighted average of the entropies of the child nodes. Entropy is a measure of randomness and is calculated using the formula: Entropy = - (probability of class 1) * log(probability of class 1) - (probability of class 2) * log(probability of class 2) - ...

Split Evaluation:

The split that results in the highest Information Gain is chosen as the best split.

3. Attribute Selection Measures:

Purpose:

These measures help determine which attribute to use for splitting at each node.

Examples:

Gini Impurity and Information Gain are commonly used attribute selection measures.

How they work:

Each measure assigns a score to each attribute, reflecting how well it separates the data. The attribute with the best score is selected as the splitting attribute.

4. Decision Tree Construction:

Recursive Process: The decision tree construction is a recursive process. At each node, the algorithm selects the best split based on the chosen attribute selection measure (e.g., Gini Impurity, Information Gain).
Splitting: The node is then split into child nodes based on the chosen split.

Termination: The splitting continues until a stopping condition is met, such as reaching a maximum tree depth, a minimum number of samples in a node, or achieving perfect purity (no more splits possible).
In essence, Decision Trees use mathematical techniques to find the most effective way to partition the data into subsets, based on the principle of maximizing homogeneity (minimizing impurity) or maximizing the reduction in randomness (maximizing information gain) within the tree structure.



8] What is Pre-Pruning in Decision Trees?

Pre-pruning (a.k.a. early stopping) is the technique of stopping the tree growth early ‚Äî before it perfectly fits the data ‚Äî to prevent overfitting.

Methods:

Maximum tree depth (max_depth)

Minimum samples per split (min_samples_split)

Minimum samples at leaf (min_samples_leaf)

Minimum impurity decrease

Goal: Stop growing the tree when further splits don‚Äôt offer significant gain.

9] What is Post-Pruning in Decision Trees?

Post-pruning is the technique of growing the full tree first, and then cutting back nodes that do not contribute significantly to model accuracy.

Steps:

Grow full tree

Evaluate each subtree (using cross-validation or validation set)

Remove branches that do not improve performance

Common method: Cost-complexity pruning (used in CART)

10] What is the difference between Pre-Pruning and Post-Pruning?

In decision trees, pre-pruning stops tree growth early, preventing it from fully developing, while post-pruning allows the tree to grow fully before removing nodes or subtrees. Pre-pruning is often referred to as "early stopping", while post-pruning is also known as "reduced error pruning".
Pre-Pruning (Early Stopping):
Process:
Pre-pruning involves setting parameters (hyperparameters) that control how large the tree can grow. The tree stops expanding when these constraints are met, such as a maximum depth or a minimum number of samples required to split a node.
Advantages:
Faster training, and can sometimes achieve similar or even better results than post-pruning.
Disadvantages:
May miss capturing complex patterns in the data, and might not prune as effectively as post-pruning.
Post-Pruning (Reduced Error Pruning):
Process:
The tree is allowed to grow fully, and then pruned by removing nodes or subtrees that don't contribute significantly to accuracy.
Advantages:
Allows for a more thorough evaluation of the tree's structure, leading to potentially more informed pruning decisions.
Disadvantages:
Slower training due to the need to evaluate the fully grown tree, and can be computationally expensive on large datasets.

11] What is a Decision Tree Regressor?

A Decision Tree Regressor is a tree-based model used for predicting continuous values instead of classes.

Splitting criterion: Variance reduction (e.g., Mean Squared Error)

Leaf node contains: Mean value of target in that region

Works similarly to classification trees but optimizes for prediction accuracy of numerical values



12] What are the advantages and disadvantages of Decision Trees?

Advantages:

Easy to interpret and visualize

No need for feature scaling

Handles both numerical and categorical data

Can capture non-linear relationships

Disadvantages:

Highly prone to overfitting

Unstable to small data changes

Greedy algorithm ‚Äî may not find global optimum

Less accurate than ensemble methods



13] How does a Decision Tree handle missing values?
Strategies:

Ignore missing values during split calculation

Imputation: Fill missing values using mean/mode/median

Surrogate Splits: Use alternative features to split missing values

Assign probabilistically: Based on distribution in other branches

Libraries like scikit-learn require missing values to be handled before training.



14] How does a Decision Tree handle categorical features?

For low-cardinality categories: One-hot encoding or label encoding works well.

For ordinal categories: Use the natural order or integer mapping.

Advanced implementations (like in LightGBM) natively support categorical splits without encoding.

In scikit-learn: You must convert categories to numbers manually via encoding.

````python
#16. Train a Decision Tree Classifier on the Iris dataset and print model accuracy

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
````

````python
#17. Train Decision Tree using Gini and print feature importances
clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(X_train, y_train)

print("Feature Importances:", clf.feature_importances_)
````

````python
#18. Train Decision Tree using Entropy and print model accuracy
clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
````

````python
#19. Train Decision Tree Regressor on housing dataset and evaluate using MSE
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

data = fetch_california_housing()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)

regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
````

````python
#21. Train tree with max_depth=3 and compare with full tree
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Train trees
shallow_tree = DecisionTreeClassifier(max_depth=3)
full_tree = DecisionTreeClassifier()

shallow_tree.fit(X_train, y_train)
full_tree.fit(X_train, y_train)

# Compare accuracy
shallow_acc = accuracy_score(y_test, shallow_tree.predict(X_test))
full_acc = accuracy_score(y_test, full_tree.predict(X_test))

print(f"Shallow tree (max_depth=3) accuracy: {shallow_acc:.4f}")
print(f"Full tree accuracy: {full_acc:.4f}")
````

````python
#22. Train with min_samples_split=5 and compare with default
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

default_tree = DecisionTreeClassifier()
min_samples_tree = DecisionTreeClassifier(min_samples_split=5)

default_tree.fit(X_train, y_train)
min_samples_tree.fit(X_train, y_train)

default_acc = accuracy_score(y_test, default_tree.predict(X_test))
min_samples_acc = accuracy_score(y_test, min_samples_tree.predict(X_test))

print(f"Default tree accuracy: {default_acc:.4f}")
print(f"min_samples_split=5 tree accuracy: {min_samples_acc:.4f}")
````

````python
#23. Apply feature scaling before training and compare
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
unscaled_tree = DecisionTreeClassifier()
scaled_tree = DecisionTreeClassifier()
unscaled_tree.fit(X_train, y_train)
scaled_tree.fit(X_train_scaled, y_train)

unscaled_acc = accuracy_score(y_test, unscaled_tree.predict(X_test))
scaled_acc = accuracy_score(y_test, scaled_tree.predict(X_test_scaled))

print(f"Unscaled data accuracy: {unscaled_acc:.4f}")
print(f"Scaled data accuracy: {scaled_acc:.4f}")
````

````python
#24. Train Decision Tree with One-vs-Rest for multiclass classification
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

ovr_clf = OneVsRestClassifier(DecisionTreeClassifier())
ovr_clf.fit(X_train, y_train)

ovr_acc = accuracy_score(y_test, ovr_clf.predict(X_test))
print(f"OvR Decision Tree accuracy: {ovr_acc:.4f}")
````

````python
#27. Train Decision Tree, apply Cost Complexity Pruning (CCP), and visualize accuracy

from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

data = load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

clf = DecisionTreeClassifier(random_state=42)
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas

accuracies = []
for alpha in ccp_alphas:
    tree = DecisionTreeClassifier(random_state=42, ccp_alpha=alpha)
    tree.fit(X_train, y_train)
    accuracies.append(accuracy_score(y_test, tree.predict(X_test)))

plt.plot(ccp_alphas, accuracies)
plt.xlabel('CCP Alpha')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Cost Complexity Pruning')
plt.show()
````

````python
#28. Evaluate using Precision, Recall, and F1-Score

from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score

data = load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
````

````python
#29. Visualize confusion matrix using seaborn

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=iris.target_names,
            yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
````

````python
#30. Use GridSearchCV to optimize max_depth and min_samples_split

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)


param_grid = {
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(
    DecisionTreeClassifier(),
    param_grid,
    cv=5,
    scoring='accuracy'
)


grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_clf = grid_search.best_estimator_

y_pred = best_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Best parameters: {best_params}")
print(f"Best model accuracy: {accuracy:.4f}")
````
