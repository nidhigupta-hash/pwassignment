Q11) What is method overloading in python?

Method overloading in Python refers to the ability to define multiple methods with the same name within a class, but with different parameter lists. Unlike some other programming languages, Python does not support traditional method overloading directly through distinct method signatures. Instead, it achieves similar functionality through techniques like default arguments, variable-length arguments, and conditional logic within a single method definition. This allows a single method to handle different numbers or types of arguments, providing flexibility and reducing code duplication.

    class Calculator:
    def add(self, a, b=None, c=None):
        if b is not None and c is not None:
            return a + b + c
        elif b is not None:
            return a + b
        else:
            return a

    calc = Calculator()
    print(calc.add(5))
    print(calc.add(5, 10))
    print(calc.add(5, 10, 15))

In this example, the add method is designed to handle one, two, or three arguments. When called with one argument, it simply returns that value. When called with two or three arguments, it returns their sum. This demonstrates how method overloading can be simulated in Python using default arguments and conditional logic.



Q12) What is method overriding in oop?

Method overriding in Python is a feature of object-oriented programming that allows a subclass to provide a specific implementation for a method that is already defined in its superclass. When a method in a subclass has the same name and signature (number and type of arguments) as a method in its superclass, the subclass's method overrides the superclass's method. This means that when the method is called on an object of the subclass, the subclass's version of the method is executed, rather than the superclass's version.

    class Animal:
        def speak(self):
            print("Generic animal sound")

    class Dog(Animal):
        def speak(self):
            print("Woof!")

    class Cat(Animal):
        def speak(self):
            print("Meow!")

    animal = Animal()
    dog = Dog()
    cat = Cat()

    animal.speak()  Output: Generic animal sound
    dog.speak()     Output: Woof!
    cat.speak()     Output: Meow!

In this example, the Animal class has a speak() method that prints a generic animal sound. The Dog and Cat classes are subclasses of Animal, and they both override the speak() method with their own specific implementations. When the speak() method is called on an object of the Dog class, the Dog class's version of the method is executed, and when it's called on an object of the Cat class, the Cat class's version is executed.

Method overriding is a powerful tool for creating flexible and reusable code. It allows subclasses to customize the behavior of their superclasses while still maintaining a consistent interface. This makes it easier to write code that can work with objects of different classes in a uniform way.

Q13) what is property decorator in python?

In Python, the @property decorator is used to define a method as a read-only property in a class. It allows a method to be accessed like an attribute, without explicitly calling it as a method.

It allows controlled access to private attributes.

It helps in implementing encapsulation (restricting direct modification of attributes).

It provides a way to define computed properties (attributes derived from other attributes).

It makes the code more Pythonic and readable.

    class Circle:
    def __init__(self, radius):
        self._radius = radius  # Private attribute

    @property
    def radius(self):
        """Getter for radius"""
        return self._radius
    
    @radius.setter
    def radius(self, value):
        """Setter for radius"""
        if value < 0:
            raise ValueError("Radius cannot be negative")
        self._radius = value
    
    @radius.deleter
    def radius(self):
        """Deleter for radius"""
        print("Deleting radius")
        del self._radius

    c = Circle(10)
    print(c.radius)  # Calls the getter
    c.radius = 5     # Calls the setter
    del c.radius     # Calls the deleter

@property makes radius behave like an attribute but calls the method behind the scenes.
@radius.setter allows setting a new value with validation.
@radius.deleter allows deleting the attribute with custom logic.

Q15) What is an abstract class in python?

An abstract class in Python is a class that cannot be instantiated directly and serves as a blueprint for other classes. It defines a common interface for its subclasses, ensuring they implement specific methods. Abstract classes are created using the abc module and the @abstractmethod decorator. They enforce a structure while allowing derived classes to provide concrete implementations for the abstract methods.

    from abc import ABC, abstractmethod

    class Shape(ABC):
        @abstractmethod
        def area(self):
            pass

        @abstractmethod
        def perimeter(self):
            pass

    class Circle(Shape):
        def __init__(self, radius):
            self.radius = radius

        def area(self):
            return 3.14 * self.radius * self.radius

        def perimeter(self):
            return 2 * 3.14 * self.radius

    class Square(Shape):
        def __init__(self, side):
            self.side = side

        def area(self):
            return self.side * self.side

        def perimeter(self):
            return 4 * self.side

    Attempting to instantiate the abstract class will raise an error

    shape = Shape()  TypeError: Can't instantiate abstract class Shape with abstract methods area, perimeter

    circle = Circle(5)
    print(f"Circle area: {circle.area()}")
    print(f"Circle perimeter: {circle.perimeter()}")

    square = Square(4)
    print(f"Square area: {square.area()}")
    print(f"Square perimeter: {square.perimeter()}")

In this example, Shape is an abstract class with abstract methods area and perimeter. Circle and Square are concrete subclasses that implement these methods. Attempting to instantiate Shape directly would result in a TypeError

Q20) what is the purpose of super function in python?

            self.name = name

                super().__init__(name)

Q21) What is the significance of del method in python?

The del statement in Python serves to remove objects, or references to objects, from namespaces. It can be used to delete variables, items from lists or dictionaries, or even slices of lists. The del statement effectively unbinds a name from an object, and if that object has no other references, it becomes eligible for garbage collection, freeing up memory.

When applied to mutable data structures like lists and dictionaries, del can modify them in place. For instance, del my_list[0] removes the first element of my_list, while del my_dict['key'] removes the key-value pair associated with 'key' from my_dict. It's important to note that del does not delete the object itself unless it's the last reference to it.

In the context of classes, del can be used to trigger the __del__ method (destructor) of an object when it's being garbage collected, allowing for cleanup actions.

Q22) What is the difference between @staticmethod and @classmethod in python?

In Python, both @staticmethod and @classmethod are decorators used to define methods within a class, but they serve different purposes and have distinct behaviors. Here's a breakdown of the differences:

1. @staticmethod
Purpose: A static method is a method that belongs to the class but does not operate on an instance or the class itself. It is essentially a regular function that is defined inside a class for organizational purposes.

Parameters: A static method does not take any special first parameter (i.e., it does not take self or cls).

Usage: It is used when the method does not need to access or modify the class or instance state.

Invocation: Can be called on both the class and an instance, but it behaves the same in both cases.


2. @classmethod
Purpose: A class method is a method that operates on the class itself rather than on instances of the class. It can modify class state that applies across all instances of the class.

Parameters: A class method takes the class as its first parameter, conventionally named cls. This allows the method to access and modify class-level attributes.

Usage: It is used when the method needs to interact with the class itself, such as when you want to create factory methods or modify class-level data.

Invocation: Can be called on both the class and an instance, but it always receives the class as the first argument.

Key Differences:
First Parameter:

@staticmethod: No special first parameter.

@classmethod: Takes cls as the first parameter, which refers to the class.

Access to Class/Instance State:

@staticmethod: Cannot access or modify class or instance state.

@classmethod: Can access and modify class state but not instance state.

Use Case:

@staticmethod: Used for utility functions that do not depend on class or instance state.

@classmethod: Used for methods that need to interact with the class itself, such as factory methods or class-level configuration.

Example Combining Both:

    class MyClass:
    class_attribute = 0

    @staticmethod
    def static_method(x, y):
        return x + y

    @classmethod
    def class_method(cls, x):
        cls.class_attribute += x
        return cls.class_attribute

    Using static method
    print(MyClass.static_method(1, 2))   Output: 3

    Using class method
    print(MyClass.class_method(5))   Output: 5
    print(MyClass.class_method(5))   Output: 10

Q23) How Polymorphism Works with Inheritance in Python?


Polymorphism in Python, particularly with inheritance, allows objects of different classes to be treated as objects of a common superclass. It enables methods to behave differently based on the object that invokes them. This is a key feature of object-oriented programming (OOP) that promotes flexibility and reusability.

Method Overriding:

When a subclass provides a specific implementation of a method that is already defined in its superclass, it is called method overriding.

The overridden method in the subclass is called instead of the method in the superclass when the method is invoked on an instance of the subclass.

Dynamic Method Dispatch:

Python determines which method to call at runtime based on the object's type. This is known as dynamic method dispatch.

This allows a method to behave differently depending on the object that invokes it.

Q24) What is method chaining in python oop?

Method chaining in Python OOP is a design pattern where multiple methods are called sequentially on an object in a single statement. Each method returns the object itself (typically self), allowing subsequent methods to be called on the result of the previous one. This creates a fluent, readable interface.

How Method Chaining Works

Return self: Each method modifies the object's state (if needed) and returns self (the instance of the class).

Chainable calls: Because each method returns the object, you can immediately call another method on the result.

        class Pizza:
            def __init__(self):
                self.size = "medium"
                self.toppings = []
            
            def set_size(self, size):
                self.size = size
                return self   Return self for chaining
            
            def add_topping(self, topping):
                self.toppings.append(topping)
                return self   Return self for chaining
            
            def bake(self):
                print(f"Baking {self.size} pizza with toppings: {', '.join(self.toppings)}")
                return self   Optional: Return self if further chaining is needed

         Using method chaining:
        pizza = (
            Pizza()
            .set_size("large")
            .add_topping("cheese")
            .add_topping("pepperoni")
            .bake()
        )

         Output:
         Baking large pizza with toppings: cheese, pepperoni

Key Features

Readability: Chains mimic natural language (e.g., obj.method1().method2().method3()).

Fluent Interface: Useful for configuring objects step-by-step (e.g., building queries, configuring settings).

State Modification: Methods can modify the objectâ€™s internal state before passing it to the next method.

Common Use Cases
Builder Patterns: Configuring complex objects (e.g., database queries, HTML generators).

Data Transformations: Libraries like pandas use chaining for operations like df.filter().groupby().sum().

Fluent APIs: Libraries like SQLAlchemy (e.g., session.query(User).filter(User.age > 25).all()).

Q25) What is the purpose of the __call__ method in python?

The __call__ method in Python allows instances of a class to be called like functions. When defined, it enables an object to be invoked using parentheses () and arguments, just like a regular function. This is useful for creating stateful callable objects or implementing function-like behavior within classes.

Key Purposes of __call__:
Make Objects Callable:
Instances can be treated as functions, allowing syntax like obj(arg1, arg2).

Maintain State Between Calls:
The object retains its internal state across multiple calls (unlike regular functions, which donâ€™t inherently preserve state).

Create Functors (Function Objects):
Useful for objects that need to behave like functions but have additional logic or data.

    class Adder:
    def __init__(self, n):
        self.n = n  # State: the number to add
    
    def __call__(self, x):
        return self.n + x  # Called like a function

    add5 = Adder(5)
    print(add5(3))   Output: 8 (5 + 3)


Benefits:

Flexibility: Objects can encapsulate both data and behavior, acting like enhanced functions.

Decorators: Implement decorators as classes.

Functors: Use in libraries like PyTorch or TensorFlow, where objects (e.g., neural network layers) are called during forward passes.

The __call__ method bridges the gap between objects and functions, letting objects retain state while behaving like functions. Itâ€™s a powerful tool for creating clean, reusable, and stateful abstractions.

```python
#Q5) Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.


class BankAccount:
    def __init__(self, initial_balance=0):
        # Private attribute
        self.__balance = initial_balance

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient balance.")

    # Method to check balance
    def check_balance(self):
        print(f"Current balance: ${self.__balance}")


account = BankAccount(initial_balance=100)

account.check_balance()

account.deposit(50)


account.withdraw(30)


account.withdraw(200)


account.deposit(-10)


account.check_balance()
```

```python
#Q14) Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary

class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    # Method to calculate the salary
    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        # Call the constructor of the base class (Employee)
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):

        base_salary = super().calculate_salary()

        return base_salary + self.bonus

manager = Manager("Alhasan", 5, 2000, 5000)

print(manager.calculate_salary())
```
