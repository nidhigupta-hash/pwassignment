Q2) How does broadcasting work in NumPy?

Broadcasting in NumPy is a powerful mechanism that allows NumPy to work with arrays of different shapes when performing arithmetic operations. The term "broadcasting" refers to how NumPy treats arrays with different shapes during arithmetic operations. The smaller array is "broadcast" across the larger array so that they have compatible shapes.

Rules of Broadcasting
For broadcasting to work, the following rules must be satisfied:

Shape Alignment:

If the arrays do not have the same number of dimensions, the shape of the smaller array is padded with ones on its left side until the shapes have the same length.

Size Compatibility:

For each dimension, the sizes must either be equal, or one of them must be 1.

If the sizes are not equal and neither is 1, a ValueError is raised.

Broadcasting Operation:

The array with size 1 in a particular dimension is stretched or "broadcast" to match the size of the other array in that dimension.

Example 1: Broadcasting a Scalar to an Array

    import numpy as np

    a = np.array([1, 2, 3])
    b = 2

    # The scalar `b` is broadcast to the shape of `a`
    result = a + b
    print(result)  # Output: [3 4 5]

Here, the scalar b is treated as if it were an array of the same shape as a, with all elements equal to 2.

Example 2: Broadcasting a 1D Array to a 2D Array

    a = np.array([[1, 2, 3], [4, 5, 6]])
    b = np.array([10, 20, 30])

    # The 1D array `b` is broadcast to match the shape of `a`
    result = a + b
    print(result)
    # Output:
    # [[11 22 33]
    #  [14 25 36]]

    In this case, the 1D array b is broadcast across the rows of the 2D array a.

Example 3: Broadcasting with Different Dimensions

    a = np.array([[1], [2], [3]])
    b = np.array([10, 20, 30])

    # The 1D array `b` is broadcast to match the shape of `a`
    result = a + b
    print(result)
    # Output:
    # [[11 21 31]
    #  [12 22 32]
    #  [13 23 33]]

Here, a is a 3x1 array, and b is a 1x3 array. Both arrays are broadcast to a 3x3 shape before the addition is performed.

Example 4: Incompatible Shapes

    a = np.array([1, 2, 3])
    b = np.array([10, 20])

    # This will raise a ValueError because the shapes are not compatible
    result = a + b

In this case, the shapes (3,) and (2,) are not compatible for broadcasting, so NumPy raises a ValueError.

Summary
Broadcasting allows NumPy to perform element-wise operations on arrays of different shapes.

The smaller array is "stretched" to match the shape of the larger array.

Broadcasting follows specific rules to ensure that the shapes are compatible.

Understanding broadcasting is crucial for efficiently working with NumPy, as it allows you to write concise and efficient code without explicitly looping over arrays.

Q3) What is a Pandas DataFrame?

A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It is one of the most commonly used data structures in Python for data manipulation and analysis. Think of it as a spreadsheet or SQL table, but with more powerful and flexible functionality.

Key Features of a DataFrame:
Tabular Structure:

Data is organized in rows and columns, similar to a table.

Each column can have a different data type (e.g., integers, floats, strings, etc.).

Labeled Axes:

Rows and columns have labels (indices).

Rows are indexed by a row label (default is integers starting from 0).

Columns are labeled by column names.

Size-Mutable:

Columns can be added or removed.

Rows can be added or deleted.

Heterogeneous Data:

Different columns can contain different data types (e.g., one column can be integers, while another can be strings).

Powerful Operations:

Supports operations like filtering, grouping, merging, reshaping, and aggregating data.

Integrates well with other Python libraries like NumPy, Matplotlib, and Scikit-learn.

Structure of a DataFrame
A DataFrame can be thought of as a dictionary of Series (where each Series represents a column). For example:

Index	Column 1	Column 2	Column 3
0	    Value 1	    Value 2	    Value 3
1	    Value 4	    Value 5	    Value 6

Index: The row labels (default is 0, 1, 2, ...).

Columns: The column labels (e.g., Column 1, Column 2, etc.).

Values: The actual data in the table.


Creating a DataFrame
You can create a DataFrame from various data sources, such as:

From a Dictionary:

Q4) Explain the use of the groupby() method in Pandas

The groupby() method in Pandas is a powerful and versatile function used to split data into groups based on some criteria, apply a function to each group, and combine the results into a new data structure. It is particularly useful for aggregating, transforming, or filtering data based on group-level operations.

Key Concepts of groupby()
Split:

The data is split into groups based on one or more keys (e.g., column values).

Apply:

A function (e.g., aggregation, transformation, or filtering) is applied to each group.

Combine:

The results from each group are combined into a new data structure (e.g., a DataFrame or Series).

Syntax

    df.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=False, observed=False, dropna=True)

by: Specifies the column(s) or key(s) to group by. This can be a column name, a list of column names, or a function.

axis: Specifies whether to group by rows (axis=0) or columns (axis=1). Default is 0.

as_index: If True, the grouped columns become the index of the resulting DataFrame. Default is True.

sort: If True, sorts the group keys. Default is True.

dropna: If True, excludes missing values in the grouping keys. Default is True.

Common Use Cases of groupby()

1. Aggregation
Aggregation involves computing a summary statistic (e.g., sum, mean, count) for each group.

    import pandas as pd

    Example DataFrame

    data = {
        'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles', 'Chicago'],
        'Sales': [100, 200, 150, 300, 250, 400]
    }

    df = pd.DataFrame(data)

    Group by 'City' and calculate the total sales for each city
    grouped = df.groupby('City')['Sales'].sum()

    print(grouped)

2. Transformation

Transformation involves performing some operation on each group and returning a DataFrame with the same shape as the original.

    Normalize sales within each city

    df['Normalized Sales'] = df.groupby('City')['Sales'].transform(lambda x: x / x.sum())

    print(df)

3. Filtering

Filtering involves removing groups that do not meet a specific condition.

    Filter out cities where the total sales are less than 300

    filtered = df.groupby('City').filter(lambda x: x['Sales'].sum() > 300)

    print(filtered)

4. Iteration

You can iterate over groups to perform custom operations.

    Iterate over each group

    for city, group in df.groupby('City'):
        print(f"City: {city}")
        print(group)

5. Multiple Columns Grouping

You can group by multiple columns to create hierarchical groupings.

    Example DataFrame with an additional column

    data = {
        'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles', 'Chicago'],
        'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Sales': [100, 200, 150, 300, 250, 400]
    }
    df = pd.DataFrame(data)

    Group by 'City' and 'Category', then calculate the total sales
    
    grouped = df.groupby(['City', 'Category'])['Sales'].sum()
    print(grouped)

Q6) What are the differences between NumPy arrays and Python lists?

NumPy arrays and Python lists are both used to store collections of data, but they have significant differences in terms of functionality, performance, and use cases. Here's a detailed comparison:

1. Data Type Handling

Python Lists:

Can store elements of different data types (e.g., integers, strings, floats, etc.).

Example:

    my_list = [1, "hello", 3.14, True]
    NumPy Arrays:

Require all elements to be of the same data type (homogeneous).

Example:

    import numpy as np
    my_array = np.array([1, 2, 3, 4])  # All elements are integers

2. Performance

Python Lists:

Slower for numerical computations because they are general-purpose and not optimized for mathematical operations.

Example:

    # Adding two lists element-wise
    list1 = [1, 2, 3]
    list2 = [4, 5, 6]
    result = [x + y for x, y in zip(list1, list2)]

NumPy Arrays:

Faster for numerical computations because they are implemented in C and optimized for performance.

Example:

    import numpy as np
    array1 = np.array([1, 2, 3])
    array2 = np.array([4, 5, 6])
    result = array1 + array2  # Vectorized operation

3. Memory Efficiency

Python Lists:

Consume more memory because they store additional information (e.g., type, reference count, etc.) for each element.

Example:

    import sys
    my_list = [1, 2, 3, 4]
    print(sys.getsizeof(my_list))  # Memory usage in bytes

NumPy Arrays:

More memory-efficient because they store data in a contiguous block of memory.

Example:

    import numpy as np
    my_array = np.array([1, 2, 3, 4])
    print(my_array.nbytes)  # Memory usage in bytes

4. Functionality

Python Lists:

Provide basic operations like appending, removing, and slicing.

Example:

    my_list = [1, 2, 3]
    my_list.append(4)  # Append an element
    my_list.remove(2)  # Remove an element

NumPy Arrays:

Offer advanced mathematical and statistical operations (e.g., matrix multiplication, Fourier transforms, etc.).

Example:

    import numpy as np
    my_array = np.array([1, 2, 3])
    print(np.mean(my_array))  # Compute the mean
    print(np.dot(my_array, my_array))  # Dot product

5. Size and Shape

Python Lists:

Do not have a fixed size or shape. They can grow or shrink dynamically.

Example:

    my_list = [1, 2, 3]
    my_list.append(4)  # Size increases

NumPy Arrays:

Have a fixed size and shape (e.g., 1D, 2D, etc.). Resizing requires creating a new array.

Example:

    import numpy as np
    my_array = np.array([1, 2, 3])
    print(my_array.shape)  # Shape of the array

6. Broadcasting

Python Lists:

Do not support broadcasting. Operations must be explicitly implemented.

Example:

    list1 = [1, 2, 3]
    list2 = [x * 2 for x in list1]  # Multiply each element by 2

NumPy Arrays:

Support broadcasting, which allows operations between arrays of different shapes.

Example:

    import numpy as np
    array1 = np.array([1, 2, 3])
    array2 = array1 * 2  # Multiply each element by 2

7. Slicing and Indexing

Python Lists:

Support basic slicing and indexing.

Example:

    my_list = [1, 2, 3, 4]
    print(my_list[1:3])  # Slice from index 1 to 2

NumPy Arrays:

Support advanced slicing and indexing (e.g., boolean indexing, integer indexing).

Example:

    import numpy as np
    my_array = np.array([1, 2, 3, 4])
    print(my_array[my_array > 2])  # Boolean indexing

8. Use Cases

Python Lists:

Ideal for general-purpose tasks, such as storing heterogeneous data or when dynamic resizing is needed.

Example: Storing a collection of user inputs or mixed data types.

NumPy Arrays:

Ideal for numerical computations, scientific computing, and working with large datasets.

Example: Performing matrix operations, statistical analysis, or machine learning tasks.

Q7) What is a heatmap, and when should it be used?

A heatmap is a graphical representation of data where individual values are represented by colors. It is a powerful visualization tool used to display the magnitude of a phenomenon as color in two dimensions. Heatmaps are particularly useful for:

Identifying Patterns and Trends: They help reveal patterns, correlations, or anomalies in data.

Comparing Values: They allow for quick comparison of values across different categories or dimensions.

Simplifying Complex Data: They make it easier to interpret large and complex datasets by summarizing them visually.

Structure of a Heatmap

A heatmap typically consists of:

Rows and Columns: Representing categories or dimensions of the data.

Cells: Each cell corresponds to a value in the dataset.

Color Gradient: Colors represent the magnitude of the values (e.g., low values in cool colors like blue, high values in warm colors like red).

When to Use a Heatmap

Heatmaps are most useful in the following scenarios:

1. Correlation Analysis

To visualize the correlation between variables in a dataset.

Example: Correlation matrix of features in a dataset.

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Load dataset
    df = sns.load_dataset('iris')

    # Compute correlation matrix
    corr = df.corr()

    # Create heatmap
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.show()

2. Time Series Data
To show how values change over time.

Example: Daily temperature variations over a year.

    import seaborn as sns
    import matplotlib.pyplot as plt
    import pandas as pd

    # Example time series data
    dates = pd.date_range(start='2023-01-01', periods=365)
    temperatures = np.random.normal(20, 5, 365)  # Random temperatures
    data = pd.DataFrame({'Date': dates, 'Temperature': temperatures})

    # Pivot data for heatmap
    data['Month'] = data['Date'].dt.month
    data['Day'] = data['Date'].dt.day
    heatmap_data = data.pivot('Day', 'Month', 'Temperature')

    # Create heatmap
    sns.heatmap(heatmap_data, cmap='coolwarm')
    plt.show()

3. Geographical Data
To represent data across geographical regions.

Example: Population density across countries.

4. Matrix Data
To visualize matrices, such as confusion matrices in machine learning.

    from sklearn.metrics import confusion_matrix
    import seaborn as sns
    import matplotlib.pyplot as plt

    # Example confusion matrix
    y_true = [0, 1, 2, 2, 0, 1]
    y_pred = [0, 2, 2, 2, 0, 0]
    cm = confusion_matrix(y_true, y_pred)

    # Create heatmap
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.show()

5. Categorical Data
To compare values across categories.

Example: Sales performance across different regions and product categories.

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Example categorical data
    data = {
        'Region': ['North', 'South', 'East', 'West'],
        'Product A': [100, 150, 200, 250],
        'Product B': [200, 250, 300, 350],
        'Product C': [300, 350, 400, 450]
    }
    df = pd.DataFrame(data).set_index('Region')

    # Create heatmap
    sns.heatmap(df, annot=True, cmap='YlGnBu')
    plt.show()

Key Features of a Heatmap

Color Gradient:

The choice of color gradient (e.g., coolwarm, viridis, YlGnBu) can highlight patterns effectively.

Example: Use a diverging colormap (e.g., coolwarm) for data with a meaningful midpoint (e.g., zero).

Annotations:

Adding numerical values to cells (annot=True) can make the heatmap more informative.

Normalization:

Normalizing data (e.g., scaling to [0, 1]) can improve the readability of the heatmap.

Clustering:

Hierarchical clustering can group similar rows or columns together to reveal patterns.

    sns.clustermap(df, cmap='viridis', annot=True)

Tools for Creating Heatmaps

Seaborn: Simplifies heatmap creation with functions like sns.heatmap() and sns.clustermap().

Matplotlib: Provides more customization options but requires more code.

Plotly: Creates interactive heatmaps for web-based visualizations.

Pandas: Can create basic heatmaps using df.style.background_gradient().

Example: Creating a Heatmap with Seaborn

    import matplotlib.pyplot as plt
    import numpy as np

    # Create a random dataset

    plt.show()

Summary

A heatmap is a versatile visualization tool used to:

Identify patterns and trends in data.

Compare values across categories or dimensions.

Simplify complex datasets by summarizing them visually.

Use heatmaps when:

You need to visualize correlations, distributions, or relationships in data.

You want to highlight variations or anomalies in a dataset.

You are working with matrix-like or tabular data.

Heatmaps are widely used in fields like data science, biology, finance, and geography to make data more accessible and interpretable.



Q8) What does the term “vectorized operation” mean in NumPy?


The term "vectorized operation" in NumPy refers to performing operations on entire arrays (or slices of arrays) at once, rather than using explicit loops to iterate over individual elements. This approach leverages optimized, low-level implementations (often written in C) to perform computations efficiently and concisely.

Key Characteristics of Vectorized Operations

Element-Wise Operations:

Operations are applied to each element of the array without the need for explicit loops.

Example: Adding two arrays element-wise.

Broadcasting:

NumPy automatically handles arrays of different shapes by "broadcasting" the smaller array to match the shape of the larger array.

Example: Adding a scalar to an array.

Performance:

Vectorized operations are highly optimized and significantly faster than Python loops, especially for large datasets.

Simplicity:

Code is more concise and readable compared to using loops.

Example: Vectorized vs Non-Vectorized Operations

Non-Vectorized (Using Loops)

    import numpy as np

    # Create two arrays
    a = np.array([1, 2, 3, 4])
    b = np.array([5, 6, 7, 8])

    # Add arrays using a loop
    result = np.empty_like(a)  # Create an empty array to store results
    for i in range(len(a)):
        result[i] = a[i] + b[i]

    print(result)  # Output: [ 6  8 10 12]

Vectorized (Using NumPy)

    import numpy as np

    # Create two arrays
    a = np.array([1, 2, 3, 4])
    b = np.array([5, 6, 7, 8])

    # Add arrays using vectorized operation
    result = a + b

    print(result)  # Output: [ 6  8 10 12]

In the vectorized version, the addition operation is applied to the entire array at once, making the code cleaner and faster.

Benefits of Vectorized Operations
Speed:

Vectorized operations are executed in optimized, pre-compiled C code, making them much faster than Python loops.

Readability:

Code is more concise and easier to understand.

Efficiency:

Reduces the need for explicit loops, which can be error-prone and inefficient.

Broadcasting:

Simplifies operations on arrays of different shapes.

Common Vectorized Operations in NumPy

1. Arithmetic Operations

Addition (+), subtraction (-), multiplication (*), division (/), etc.

    a = np.array([1, 2, 3])
    b = np.array([4, 5, 6])
    result = a * b  # Element-wise multiplication
    print(result)  # Output: [ 4 10 18]

2. Mathematical Functions

Trigonometric functions (np.sin, np.cos), logarithms (np.log), exponentials (np.exp), etc.

    a = np.array([1, 2, 3])
    result = np.exp(a)  # Element-wise exponential
    print(result)  # Output: [ 2.71828183  7.3890561  20.08553692]

3. Comparison Operations

Greater than (>), less than (<), equal to (==), etc.

    a = np.array([1, 2, 3])
    b = np.array([2, 2, 2])
    result = a > b  # Element-wise comparison
    print(result)  # Output: [False False  True]

4. Aggregation Functions

Sum (np.sum), mean (np.mean), max (np.max), min (np.min), etc.

    a = np.array([1, 2, 3, 4])
    result = np.sum(a)  # Sum of all elements
    print(result)  # Output: 10

5. Logical Operations

Logical AND (&), OR (|), NOT (~), etc.

    a = np.array([True, False, True])
    b = np.array([False, False, True])
    result = a & b  # Element-wise logical AND
    print(result)  # Output: [False False  True]
6. Broadcasting

Operations between arrays of different shapes.

    a = np.array([1, 2, 3])
    b = 2
    result = a * b  # Scalar broadcasted to array
    print(result)  # Output: [2 4 6]

Why Use Vectorized Operations?

Performance:

Vectorized operations are optimized for speed, making them ideal for large-scale computations.

Simplicity:

Code is cleaner and easier to maintain.

Integration:

NumPy integrates well with other scientific computing libraries like SciPy, Pandas, and Scikit-learn.

Parallelism:

Many vectorized operations can be parallelized, taking advantage of modern hardware (e.g., multi-core processors).


Non-Vectorized Approach

    celsius = [0, 10, 20, 30, 40]
    fahrenheit = []

    df['Age'] = df['Age'].interpolate()  # Fills missing values using linear interpolation

np.ones(): Creates an array filled with ones.

NumPy arrays have attributes that describe their properties:

shape: Dimensions of the array (e.g., (2, 3) for a 2x3 matrix).

ndim: Number of dimensions (e.g., 2 for a 2D array).

size: Total number of elements in the array.

dtype: Data type of the array elements (e.g., int64, float32).

Q21) Why is NumPy’s array slicing faster than Python’s list slicing?

NumPy’s array slicing is significantly faster than Python’s list slicing due to several key reasons, including memory efficiency, contiguous storage, and optimized C-based operations.

1. NumPy Uses Contiguous Memory Storage
NumPy arrays are stored in a contiguous block of memory, unlike Python lists, which store references to objects in memory.
This contiguous layout allows faster access and better cache locality, improving performance.

Example:

    import numpy as np

    arr = np.array([1, 2, 3, 4, 5])
    slice_arr = arr[1:4]  # Creates a view, not a copy

Here, NumPy does not copy the data but creates a view of the same memory location.

2. NumPy Uses Views Instead of Copies

Slicing a Python list creates a new list, requiring extra time and memory.
NumPy does not copy the data; it returns a view, meaning changes affect the original array.

Example: Python List (Copying Data)

    lst = [1, 2, 3, 4, 5]
    slice_lst = lst[1:4]  # Creates a new list
    slice_lst[0] = 100  # Does not affect the original list

Example: NumPy Array (Using a View)

    arr = np.array([1, 2, 3, 4, 5])
    slice_arr = arr[1:4]  # Creates a view
    slice_arr[0] = 100  # Modifies the original array

    print(arr)  # Output: [  1 100   3   4   5]

NumPy avoids the overhead of copying by using views instead of duplicating memory.

3. NumPy is Implemented in Optimized C Code
NumPy operations are written in highly optimized C and Fortran, making them much faster than native Python loops or list operations.
Python lists require looping through each element manually in Python, which is slower.

Example: Timing Python List vs. NumPy Array Slicing

    import numpy as np
    import time

    lst = list(range(1_000_000))
    arr = np.array(lst)

    # Python list slicing
    start = time.time()
    slice_lst = lst[:500_000]
    end = time.time()
    print("Python list slicing time:", end - start)

    # NumPy array slicing
    start = time.time()
    slice_arr = arr[:500_000]
    end = time.time()
    print("NumPy array slicing time:", end - start)

NumPy slicing is significantly faster due to its optimized implementation.

4. NumPy Uses Vectorized Operations

Python lists require loops for element-wise operations, whereas NumPy performs vectorized operations using efficient C implementations.
This results in faster execution, especially for large datasets.

Example: Squaring Elements in a List vs. NumPy Array

lst = list(range(10_000))
arr = np.array(lst)

# Using Python list (loop required)
%timeit [x**2 for x in lst]

# Using NumPy (vectorized)
%timeit arr**2
NumPy executes the operation much faster due to low-level optimizations.

Conclusion

NumPy’s array slicing is faster than Python’s list slicing because:

Uses contiguous memory storage, leading to better cache performance.

Returns views instead of copies, reducing memory overhead.

Implemented in C, making operations much faster.

Supports vectorized operations, eliminating the need for slow Python loops.

This makes NumPy the preferred choice for numerical computations and large datasets.

```python
#Q4) How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 55000, 65000, 70000],
    'Experience': [1, 3, 5, 7, 9]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```

```python
#Q6) Create a DataFrame and add a new column based on an existing column

import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alhasan', 'Faisal', 'Vaibhav', 'Danish', 'Erturgul'],
    'Age': [25, 30, 35, 40, 45]
}

df = pd.DataFrame(data)

# Adding a new column 'Age Group' based on the 'Age' column
df['Age Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Middle-aged' if x < 40 else 'Senior')

# Display the DataFrame
print(df)
```

```python
#Q7) Write a program to perform element-wise multiplication of two NumPy arrays

import numpy as np

# Creating two NumPy arrays
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([10, 20, 30, 40, 50])

# Element-wise multiplication using `*` operator
result1 = array1 * array2

# Element-wise multiplication using `np.multiply()`
result2 = np.multiply(array1, array2)

# Display results
print("Array 1:", array1)
print("Array 2:", array2)
print("Result using `*` operator:", result1)
print("Result using `np.multiply()`:", result2)
```

```python
#Q8) Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]  # First line
y2 = [5, 15, 20, 25, 35]   # Second line
y3 = [2, 10, 15, 22, 28]   # Third line

# Plot multiple lines
plt.plot(x, y1, label="Line 1", marker='o', linestyle='-')
plt.plot(x, y2, label="Line 2", marker='s', linestyle='--')
plt.plot(x, y3, label="Line 3", marker='d', linestyle=':')

# Adding labels and title
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Plot with Multiple Lines")

# Display legend
plt.legend()

# Show the plot
plt.show()
```

```python
#Q9) Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alhasan', 'Faisal', 'Vaibhav', 'Danish', 'Erturgul'],
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 55000, 65000, 70000]
}

df = pd.DataFrame(data)

# Define threshold (e.g., filter salaries greater than 55,000)
threshold = 55000
filtered_df = df[df['Salary'] > threshold]

# Display the filtered DataFrame
print(filtered_df)
```

```python
#Q10) Create a histogram using Seaborn to visualize a distribution

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (Normal distribution)
data = np.random.randn(200) * 10 + 50  # Mean=50, Std=10

# Create a histogram
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=20, kde=True, color='blue')

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data Distribution")

# Show the plot
plt.show()
```

```python
#Q11) Perform matrix multiplication using NumPy.

import numpy as np

# Define two matrices
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Perform matrix multiplication
result1 = np.dot(A, B)     # Using np.dot()
result2 = A @ B            # Using @ operator
result3 = np.matmul(A, B)  # Using np.matmul()

# Display results
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Result using np.dot():\n", result1)
print("Result using @ operator:\n", result2)
print("Result using np.matmul():\n", result3)
```

```python
#Q13) Create a 3D scatter plot using Plotly

import plotly.graph_objects as go
import numpy as np

# Generate random data for 3D scatter plot
np.random.seed(42)  # For reproducibility
x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)
z = np.random.randint(1, 100, 20)

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(size=6, color=z, colorscale='Viridis', opacity=0.8)
)])

# Add title and labels
fig.update_layout(
    title="3D Scatter Plot Using Plotly",
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Show the plot
fig.show()
```
