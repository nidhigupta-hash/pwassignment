# Evaluation Metrics and Regression Implementation

1) What does R-squared represent in a regression model:

R-squared, also known as the coefficient of determination, measures the proportion of the variance in the dependent (target) variable that is explained by the independent (predictor) variables in the model. It provides a sense of how well the model fits the data.

For example, an R-squared of 0.70 means that 70% of the variance in the outcome variable is explained by the model’s predictors, and 30% remains unexplained (i.e., captured in the residuals). It ranges between 0 and 1, where 1 indicates a perfect fit, and 0 means the model does not explain any variance.

However, R-squared does not indicate whether the model is appropriate or whether predictors are statistically significant. Also, it can increase when more predictors are added, even if they are not useful — which is why we also use Adjusted R-squared.

2] What are the assumptions of linear regression:

Linear regression relies on several key assumptions to ensure that the estimated coefficients are accurate and that statistical inferences (e.g., confidence intervals, hypothesis tests) are valid:

Linearity: The relationship between the independent variables and the dependent variable should be linear. This means the effect of the predictors on the outcome should be additive and proportional.

Independence of errors: Observations should be independent of each other, and the residuals (errors) should not be correlated. Violation of this (e.g., in time series data) leads to biased estimates.

Homoscedasticity: The variance of residuals should be constant across all levels of the independent variables. If residuals fan out or contract, this is called heteroscedasticity and can distort standard errors.

Normality of residuals: The residuals (not necessarily the dependent variable) should be normally distributed for valid hypothesis testing and confidence intervals.

No multicollinearity: The independent variables should not be highly correlated with each other. High correlation among predictors can make coefficient estimates unstable and difficult to interpret.

3] What is the difference between R-squared and Adjusted R-squared:

R-squared always increases when more variables are added to the model, regardless of whether those variables improve the model significantly.

Adjusted R-squared modifies R-squared to penalize the model for including unnecessary predictors. It adjusts for the number of variables in the model relative to the number of observations.

Mathematically, Adjusted R² adds a correction for the number of predictors. If the new variable doesn’t improve the model much, Adjusted R² can actually decrease. This makes it a better metric for comparing models with different numbers of predictors.

4] Why do we use Mean Squared Error (MSE):

MSE is used as a loss function and evaluation metric in regression because it provides a single value that summarizes the average squared difference between actual and predicted values. The squaring ensures that larger errors have a disproportionately larger effect on the overall error, which can be useful when large errors are particularly undesirable.

MSE also has nice mathematical properties — it is differentiable and smooth, which is ideal for optimization algorithms like gradient descent during model training.

5] What does an Adjusted R-squared value of 0.85 indicate:

An Adjusted R-squared of 0.85 suggests that 85% of the variability in the target variable is explained by the model, after accounting for the number of predictors used. This is generally considered a very good fit, especially in fields where high variability is common. It also implies that the model is likely not overfitting, assuming assumptions are met.



6] How do we check for normality of residuals in linear regression:

To validate the assumption that residuals are normally distributed, we can:

Histogram of residuals: If the histogram resembles a bell curve, residuals are likely normal.

Q-Q plot (quantile-quantile plot): This plot compares quantiles of residuals to those of a normal distribution. If the points lie roughly on a straight line, the residuals are normal.

Statistical tests:

Shapiro-Wilk Test (sensitive for small samples),

Kolmogorov-Smirnov Test, or

Anderson-Darling Test. If the p-value is low (typically < 0.05), it suggests non-normality.

7] What is multicollinearity, and how does it impact regression:

Multicollinearity occurs when two or more independent variables are highly correlated. This causes problems because it becomes difficult to determine the individual effect of each predictor on the response variable.

Consequences include:

Unstable coefficients (small changes in data can lead to large changes in coefficients),

Inflated standard errors, leading to lower statistical significance,

Difficulty in model interpretation.

We can detect multicollinearity using:

Correlation matrix,

Variance Inflation Factor (VIF) — values >10 are problematic.

8] What is Mean Absolute Error (MAE):

MAE is the average of the absolute differences between predicted values and actual values. Unlike MSE, it does not square the errors, so it treats all deviations equally. It is more robust to outliers than MSE and gives a more straightforward interpretation — the average size of the prediction error in the same units as the target variable.

9] What are the benefits of using an ML pipeline:

An ML pipeline is a series of steps that automate the process of training a machine learning model. It typically includes preprocessing, feature transformation, model fitting, and prediction.

Benefits include:

Automation and Reproducibility: Once built, the same pipeline can be reused consistently.

Error Reduction: Reduces risk of manual errors and data leakage.

Simplified Workflow: Makes the entire model development process organized and modular.

Parameter tuning: Pipelines make cross-validation and hyperparameter tuning easier using tools like GridSearchCV.

10] Why is RMSE considered more interpretable than MSE:

RMSE (Root Mean Squared Error) is the square root of MSE, which brings the error back to the same unit as the target variable. For example, if you're predicting house prices in dollars, RMSE is also in dollars, making it easier to interpret what the error means in real-world terms. MSE, being squared, distorts the scale.

11] What is pickling in Python, and how is it useful in ML:
Pickling is a way to serialize Python objects into a byte stream so they can be saved to a file and loaded later. In ML, we often pickle:

Trained models (e.g., using pickle or joblib),

Preprocessing pipelines (like StandardScaler, PCA, etc.).

This allows us to reuse the trained model without retraining, which is crucial for deploying models to production.



12] What does a high R-squared value mean:

A high R² value indicates that a large proportion of the variation in the dependent variable is explained by the independent variables. However, it doesn't necessarily mean that the model is good — you can have a high R² with a model that overfits or includes irrelevant variables. Also, R² doesn’t detect biases or violations of assumptions.



13] What happens if linear regression assumptions are violated:

Violating assumptions leads to:

Biased or inefficient estimates,

Incorrect conclusions from p-values and confidence intervals,

Overstated or understated model performance.

For example:

Non-normal residuals affect hypothesis testing,

Heteroscedasticity leads to biased standard errors,

Multicollinearity distorts coefficient interpretation.



14] How can we address multicollinearity in regression:

Remove one of the correlated variables if it’s not essential.

Combine correlated features using dimensionality reduction (e.g., PCA).

Use regularization techniques:

Ridge regression (L2 penalty),

Lasso regression (L1 penalty) which can zero out some coefficients.

Centering the variables can sometimes reduce multicollinearity due to interactions.

15] How can feature selection improve model performance in regression analysis:

Reduces overfitting by removing irrelevant variables,

Improves interpretability since fewer variables make the model easier to understand,

Increases training efficiency by reducing computation,

Helps avoid multicollinearity and noise from redundant variables.

Methods include:

Manual selection based on domain knowledge,

Statistical tests,

Recursive Feature Elimination (RFE),

Lasso regression (which inherently selects features).

17] Why is MSE sensitive to outliers:

MSE squares the errors before averaging. So a large error (e.g., from an outlier) becomes much larger when squared. For example, an error of 10 becomes 100. This gives outliers a disproportionately large influence on the final metric, making MSE sensitive to them.



18]: What is the role of homoscedasticity in linear regression:

Homoscedasticity means that the variance of the residuals is constant across all levels of the independent variables. This is important because:

If violated (i.e., heteroscedasticity), it affects the accuracy of coefficient standard errors,

This, in turn, affects hypothesis testing (t-tests, F-tests),

It may also indicate that a transformation or different model is needed.

19] What is Root Mean Squared Error (RMSE):

RMSE is the square root of the average of the squared differences between predicted and actual values. Like MSE, it penalizes larger errors more than smaller ones, but it has the advantage of being in the same unit as the dependent variable.

20] Why is pickling considered risky:

Pickle files can execute arbitrary code when loaded. If someone tampers with the file (e.g., inserts malicious code), and you unpickle it, that code can execute and harm your system. That’s why it’s strongly advised to never unpickle data from untrusted sources.

21] What alternatives exist to pickling for saving ML models:

Joblib: More efficient for models with large NumPy arrays.

ONNX (Open Neural Network Exchange): Open standard for saving and sharing models across platforms.

HDF5 (.h5): Used for deep learning models (e.g., Keras/TensorFlow).

MLflow: For managing the full ML lifecycle, including saving models.

SavedModel (TensorFlow): TensorFlow's preferred format.

PMML: An XML-based standard to represent predictive models.

22] What is heteroscedasticity, and why is it a problem:

Heteroscedasticity means the variance of residuals is not constant across all levels of the independent variables. For example, errors may be small for low values and large for high values of a predictor.

This violates the assumption of homoscedasticity and can lead to:

Inefficient estimates,

Biased standard errors, making confidence intervals and p-values unreliable.

23] How can interaction terms enhance a regression model's predictive power:

Interaction terms allow the effect of one predictor on the target variable to depend on the value of another predictor. For example, the effect of exercise on weight loss might differ by age — this can be modeled using an interaction term between age and exercise level.

Adding interaction terms can:

Capture more complex relationships,

Improve predictive accuracy,

Make the model more realistic when variables influence each other.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats


diamonds = sns.load_dataset("diamonds")


features = ['carat', 'depth', 'table', 'x', 'y', 'z']
target = 'price'


data = diamonds[features + [target]].dropna()


X = data[features]
y = data[target]


X = sm.add_constant(X)


model = sm.OLS(y, X).fit()


residuals = model.resid


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(residuals, kde=True, color='skyblue', bins=40)
plt.title('Distribution of Residuals')
plt.xlabel('Residual')
plt.ylabel('Frequency')


plt.subplot(1, 2, 2)
sm.qqplot(residuals, line='45', fit=True, ax=plt.gca())
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
import numpy as np

data = fetch_california_housing()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("MSE:", mse)
print("MAE:", mae)
print("RMSE:", rmse)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statsmodels.api as sm

diamonds = sns.load_dataset("diamonds")
df = diamonds[['carat', 'depth', 'table', 'price']].dropna()

X = df[['carat', 'depth', 'table']]
y = df['price']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
residuals = y - predictions

sns.pairplot(df[['carat', 'depth', 'table', 'price']])
plt.suptitle("Linearity Check")
plt.show()

plt.scatter(predictions, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuals vs Fitted")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()

corr = df[['carat', 'depth', 'table']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import r2_score

X, y = fetch_california_housing(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(),
    'Random Forest': RandomForestRegressor()
}

for name, model in models.items():
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('regressor', model)
    ])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    print(f"{name} R² Score: {r2_score(y_test, y_pred):.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

X = np.linspace(0, 10, 100).reshape(-1, 1)
y = 2 * X.squeeze()**3 - 5 * X.squeeze()**2 + 3 * X.squeeze() + np.random.randn(100) * 50

degrees = [1, 2, 3, 4, 5]

for d in degrees:
    poly = PolynomialFeatures(degree=d)
    X_poly = poly.fit_transform(X)
    model = LinearRegression().fit(X_poly, y)
    y_pred = model.predict(X_poly)
    mse = mean_squared_error(y, y_pred)
    plt.plot(X, y_pred, label=f'Degree {d} (MSE={mse:.0f})')

plt.scatter(X, y, alpha=0.3)
plt.legend()
plt.title("Polynomial Regression Comparison")
plt.show()
```

```python
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

X, y = make_regression(n_samples=100, n_features=10, noise=10)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

lr = LinearRegression()
lr.fit(X_train, y_train)

ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

print("Linear Regression R²:", lr.score(X_test, y_test))
print("Ridge Regression R²:", ridge.score(X_test, y_test))
print("Linear Coefficients:", lr.coef_)
print("Ridge Coefficients:", ridge.coef_)
```
